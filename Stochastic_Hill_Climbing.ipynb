# **1) [5.0 pontos] Implemente uma versão do algoritmo Stochastic Hill Climbing, em uma linguagem de programação de sua escolha, para resolver o problema das oito rainhas. Utilize a codificação de array de 8 posições, onde cada índice representa uma coluna e cada posição armazena a linha correspondente, conforme a imagem abaixo que mostra uma solução desejada.**


image.png

```python
import random
import time
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class SHC:
  def __init__(self, iteracoes_maximas=500):
    self.tabuleiro = []
    self.iteracoes_maximas = iteracoes_maximas

  def stochastic_hill_climbing(self):
    """
    Executa o algoritmo de subida de encosta estocástica.
    """
    inicio = time.time()
    iterações_totais = 0

    while True:
      # Random Restart
      atual = [random.randint(0, 7) for _ in range(8)]
      h_atual = self.calcular_ataques(atual)

      if h_atual == 0: # solução imediata
        fim = time.time()
        self.tabuleiro = atual
        return iterações_totais, fim - inicio

      for _ in range(self.iteracoes_maximas):
        iterações_totais += 1

        vizinho = self.obter_vizinho_aleatorio(atual)
        h_vizinho = self.calcular_ataques(vizinho)

        if h_vizinho < h_atual:
          atual = vizinho
          h_atual = h_vizinho

          if h_atual == 0: # solução encontrada
            fim = time.time()
            self.tabuleiro = atual
            return iterações_totais, fim - inicio

  def calcular_ataques(self, tabuleiro):
    """
    Calcula o número de pares de rainhas se atacando (heurística h).
    Se h = 0, temos uma solução.
    """
    n = len(tabuleiro)
    ataques = 0

    for i in range(n):
      for j in range(i + 1, n):
        # Checa ataque na mesma linha (horizontal)
        if tabuleiro[i] == tabuleiro[j]:
          ataques += 1

        # Checa ataque na diagonal
        # A diferença absoluta entre as colunas é igual à diferença absoluta entre as linhas
        elif abs(tabuleiro[i] - tabuleiro[j]) == abs(i - j):
          ataques += 1

    return ataques

  def obter_vizinho_aleatorio(self, tabuleiro):
    """
    Gera um vizinho aleatório movendo uma rainha de uma coluna
    para uma nova linha aleatória.
    """
    vizinho = list(tabuleiro)

    coluna = random.randint(0, 7)
    nova_linha = random.randint(0, 7)

    # Garante que a posição mudou de fato
    while vizinho[coluna] == nova_linha:
      nova_linha = random.randint(0, 7)

    vizinho[coluna] = nova_linha
    return vizinho

  def imprimir_tabuleiro(self):
    """
    Imprime o tabuleiro de forma visual.
    """
    print(f"Array de representação: {self.tabuleiro}\n")
    print(" "+"".join(f" |{str(i)}|" for i in range(8)))
    for linha in range(8):
      row_str = f"{linha} "
      for col in range(8):
        if self.tabuleiro[col] == linha:
          row_str += "👑  "
        else:
          if abs(col - linha) & 1:
            row_str += "⬜  "
          else:
            row_str += "⬛  "
      print(row_str)
```

### Execução com Random Restart

```python
shc = SHC()
iteracoes, tempo = shc.stochastic_hill_climbing()
shc.imprimir_tabuleiro()
```

## **b) Execute o algoritmo 50 vezes e calcule: média e desvio padrão do número mínimo de iterações necessário para parar o algoritmo; média e desvio padrão do tempo de execução do algoritmo.**

```python
num_execucoes = 50
lista_iteracoes = []
lista_tempos = []

for i in range(num_execucoes):
  tabuleiro = [random.randint(0, 7) for _ in range(8)]
  shc = SHC()
  iters, tempo = shc.stochastic_hill_climbing()
  lista_iteracoes.append(iters)
  lista_tempos.append(tempo)

media_iter = statistics.mean(lista_iteracoes)
desvio_iter = statistics.stdev(lista_iteracoes)

media_tempo = statistics.mean(lista_tempos)
desvio_tempo = statistics.stdev(lista_tempos)

print("\nResultados:")
print(f"Qauntidade de execuções: {num_execucoes}\n")

print(f"Média de Iterações: {media_iter:.2f}")
print(f"Desvio Padrão das Iterações: {desvio_iter:.2f}\n")

print(f"Média de Tempo: {media_tempo:.4f} segundos")
print(f"Desvio Padrão do Tempo: {desvio_tempo:.4f} segundos")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(lista_iteracoes, kde=True, bins=20, ax=axes[0])
axes[0].set_title('Distribuição das Iterações para encontrar a solução')
axes[0].set_xlabel('Número de Iterações')
axes[0].set_ylabel('Frequência')
axes[0].grid(True)

sns.histplot(lista_tempos, kde=True, bins=20, ax=axes[1])
axes[1].set_title('Distribuição do tempo de execução para encontrar a solução')
axes[1].set_xlabel('Tempo de execução (segundos)')
axes[1].set_ylabel('Frequência')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### Análise e Resultados

Os resultados demonstram uma alta variabilidade intrínseca ao algoritmo, evidenciada pelo fato de o desvio padrão das iterações estar próximo ou superar a média. Esse comportamento ocorre porque o Stochastic Hill Climbing é uma estratégia gulosa e "míope", propensa a ficar presa em máximos locais onde nenhuma melhoria simples é possível. Como o algoritmo não possui memória, o sucesso depende fortemente da sorte na configuração inicial; enquanto algumas execuções encontram o caminho direto para a solução, outras exigem múltiplos reinícios (Random Restarts) para escapar de estados estagnados, gerando a grande dispersão observada nos dados.

Apesar dessa inconsistência no número de passos, o tempo absoluto de execução manteve-se baixo, refletindo a simplicidade computacional da função de avaliação no tabuleiro 8×8. Isso sugere que, para este tamanho de problema, a estratégia de realizar muitas tentativas rápidas é viável e eficiente. O método compensa sua falta de planejamento global e a necessidade de reinícios frequentes com a velocidade bruta de processamento, provando que a abordagem de "tentativa e erro" massiva funciona bem quando o custo de cada verificação é desprezível.

## **c) Mostre as cinco melhores soluções distintas encontradas pelo algoritmo.**

```python
solver = SHC()
solucoes_unicas = []
hashes_vistos = set()

while len(solucoes_unicas) < 5:
  solver.stochastic_hill_climbing()

  estado_tupla = tuple(solver.tabuleiro)

  if estado_tupla not in hashes_vistos:
    hashes_vistos.add(estado_tupla)
    solucoes_unicas.append(list(estado_tupla))

for i, solucao in enumerate(solucoes_unicas):
  print(f"\n--- Solução {i+1} ---")
  solver.tabuleiro = solucao
  solver.imprimir_tabuleiro()
```
