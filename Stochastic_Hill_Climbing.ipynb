#Fun√ß√µes auxiliares

```python
def calcular_ataques(tabuleiro):
    """
    Calcula o n√∫mero de pares de rainhas se atacando (heur√≠stica h).
    Se h = 0, temos uma solu√ß√£o.
    """
    n = len(tabuleiro)
    ataques = 0

    for i in range(n):
        for j in range(i + 1, n):
            # Checa ataque na mesma linha (horizontal)
            if tabuleiro[i] == tabuleiro[j]:
                ataques += 1

            # Checa ataque na diagonal
            # A diferen√ßa absoluta entre as colunas √© igual √† diferen√ßa absoluta entre as linhas
            elif abs(tabuleiro[i] - tabuleiro[j]) == abs(i - j):
                ataques += 1

    return ataques

def imprimir_tabuleiro(tabuleiro):
    """
    Imprime o tabuleiro de forma visual recebendo um array de posi√ß√µes.
    """
    if not tabuleiro:
        print("Tabuleiro vazio.")
        return

    print(f"Array de representa√ß√£o: {tabuleiro}\n")
    # Cabe√ßalho das colunas
    print("   " + " ".join(f" {i} " for i in range(8)))

    for linha in range(8):
        # N√∫mero da linha √† esquerda
        row_str = f"{linha} "
        for col in range(8):
            # Se a rainha da coluna 'col' est√° nesta 'linha'
            if tabuleiro[col] == linha:
                row_str += " üëë "
            else:
                # L√≥gica para pintar casas brancas e pretas (tabuleiro de xadrez)
                if (col + linha) % 2 == 0:
                    row_str += " ‚¨ú "
                else:
                    row_str += " ‚¨õ "
        print(row_str)
```

```python
import random
import time
import statistics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class SHC:
    def __init__(self, iteracoes_maximas=500):
        self.tabuleiro = []
        self.iteracoes_maximas = iteracoes_maximas

    def stochastic_hill_climbing(self) -> tuple[int, float]:
        """
        Executa o algoritmo de subida de encosta estoc√°stica.
        Retorna o n√∫mero de itera√ß√µes executadas e o tempo de execu√ß√£o.

        Returns
        -------
        int
            N√∫mero de itera√ß√µes executadas.
        float
            Tempo de execu√ß√£o.
        """
        inicio = time.time()
        iteracoes_totais = 0

        while True:
            # Random Restart: Come√ßa com um estado aleat√≥rio
            atual = [random.randint(0, 7) for _ in range(8)]

            h_atual = calcular_ataques(atual)

            if h_atual == 0: # solu√ß√£o imediata
                fim = time.time()
                self.tabuleiro = atual
                return iteracoes_totais, fim - inicio

            # Tenta melhorar a solu√ß√£o por um n√∫mero m√°ximo de itera√ß√µes
            for _ in range(self.iteracoes_maximas):
                iteracoes_totais += 1

                vizinho = self.obter_vizinho_aleatorio(atual)

                h_vizinho = calcular_ataques(vizinho)

                # Se o vizinho √© melhor, movemos para ele
                if h_vizinho < h_atual:
                    atual = vizinho
                    h_atual = h_vizinho

                    if h_atual == 0: # solu√ß√£o encontrada
                        fim = time.time()
                        self.tabuleiro = atual
                        return iteracoes_totais, fim - inicio

    def obter_vizinho_aleatorio(self, tabuleiro):
        """
        Gera um vizinho aleat√≥rio movendo uma rainha de uma coluna
        para uma nova linha aleat√≥ria.
        """
        vizinho = list(tabuleiro)

        coluna = random.randint(0, 7)
        nova_linha = random.randint(0, 7)

        # Garante que a posi√ß√£o mudou de fato
        while vizinho[coluna] == nova_linha:
            nova_linha = random.randint(0, 7)

        vizinho[coluna] = nova_linha
        return vizinho
```

### Execu√ß√£o do algoritmo Hill Climbing com random restart

```python
shc = SHC()
iteracoes, tempo = shc.stochastic_hill_climbing()
print(f"Execu√ß√µes realizadas: {iteracoes}\nTempo de execu√ß√£o: {tempo} segundo(s)")
imprimir_tabuleiro(shc.tabuleiro)
```

O processo inicia-se com a gera√ß√£o de um estado completamente arbitr√°rio, onde as oito rainhas s√£o distribu√≠das aleatoriamente pelo tabuleiro, seguido imediatamente pelo c√°lculo da fun√ß√£o de aptid√£o (fitness), que contabiliza o n√∫mero total de colis√µes entre as pe√ßas. Se essa configura√ß√£o inicial n√£o apresentar conflitos, o algoritmo encerra com sucesso; caso contr√°rio, entra-se em um ciclo iterativo de busca controlado por um contador de estagna√ß√£o, cujo objetivo √© monitorar se o sistema ainda est√° evoluindo ou se ficou preso.

Dentro desse ciclo, o algoritmo prop√µe um novo estado vizinho movendo uma rainha aleat√≥ria e compara o n√∫mero de colis√µes desse novo cen√°rio com o atual. Se o vizinho apresentar um fitness melhor (menos ataques), a mudan√ßa √© aceita, o estado atual √© atualizado e o contador de estagna√ß√£o √© zerado, validando o progresso. No entanto, se o vizinho n√£o oferecer melhoria, a mudan√ßa √© descartada e o contador de tentativas sem sucesso √© incrementado. Caso esse contador atinja o limite cr√≠tico de 500 itera√ß√µes consecutivas sem reduzir o n√∫mero de ataques, o algoritmo conclui que est√° preso em um m√°ximo local e aciona o mecanismo de rein√≠cio (Random Restart), descartando o tabuleiro atual e gerando uma nova configura√ß√£o aleat√≥ria para recome√ßar a busca em outra regi√£o do espa√ßo de solu√ß√µes, repetindo esse fluxo at√© que o objetivo de zero conflitos seja alcan√ßado.

Esse comportamento ocorre porque o Stochastic Hill Climbing √© uma estrat√©gia gulosa e "m√≠ope", propensa a ficar presa em m√°ximos locais onde nenhuma melhoria simples √© poss√≠vel. Como o algoritmo n√£o possui mem√≥ria, o sucesso depende fortemente da sorte na configura√ß√£o inicial; enquanto algumas execu√ß√µes encontram o caminho direto para a solu√ß√£o, outras exigem m√∫ltiplos rein√≠cios (Random Restarts) para escapar de estados estagnados, gerando a grande dispers√£o observada nos dados.

Apesar dessa inconsist√™ncia no n√∫mero de passos, o tempo absoluto de execu√ß√£o manteve-se baixo, refletindo a simplicidade computacional da fun√ß√£o de avalia√ß√£o no tabuleiro 8√ó8. O m√©todo compensa sua falta de planejamento global e a necessidade de rein√≠cios frequentes com a velocidade bruta de processamento, provando que a abordagem de "tentativa e erro" massiva funciona bem quando o custo de cada verifica√ß√£o √© desprez√≠vel.

# **2) [5.0 pontos] Desenvolva um algoritmo gen√©tico, em uma linguagem de programa√ß√£o de sua escolha, para resolver o problema das oito rainhas descrito na quest√£o anterior. Utilize obrigatoriamente a codifica√ß√£o bin√°ria para representar um indiv√≠duo. Os par√¢metros desse algoritmo gen√©tico devem ser:**

##Configura√ß√µes do problema

```python
TAMANHO_POPULACAO = 20
NUM_RAINHAS = 8
TAMANHO_CROMOSSOMO = NUM_RAINHAS * 3
TAXA_CRUZAMENTO = 0.80
TAXA_MUTACAO = 0.03
MAX_GERACOES = 1000
```

##Algoritmo gen√©tico

```python
def decodificar_binario(cromossomo):
    tabuleiro = []
    for i in range(0, len(cromossomo), 3):
        bits = cromossomo[i : i + 3]
        string_bits = "".join(str(b) for b in bits)
        numero = int(string_bits, 2)
        tabuleiro.append(numero)
    return tabuleiro

def calcular_aptidao(cromossomo):
    tabuleiro = decodificar_binario(cromossomo)
    colisoes = calcular_ataques(tabuleiro)

    return 28 - colisoes

def selecao_roleta(populacao, notas_aptidao):
    soma_aptidao = sum(notas_aptidao)
    if soma_aptidao == 0:
        return random.choice(populacao)

    sorteio = random.uniform(0, soma_aptidao)
    acumulado = 0
    for individuo, nota in zip(populacao, notas_aptidao):
        acumulado += nota
        if acumulado > sorteio:
            return individuo
    return populacao[-1]

def cruzamento(pai1, pai2):
    if random.random() < TAXA_CRUZAMENTO:
        ponto_corte = random.randint(1, TAMANHO_CROMOSSOMO - 2)
        filho1 = pai1[:ponto_corte] + pai2[ponto_corte:]
        filho2 = pai2[:ponto_corte] + pai1[ponto_corte:]
        return filho1, filho2
    return pai1[:], pai2[:]

def mutacao(individuo):
    for i in range(len(individuo)):
        if random.random() < TAXA_MUTACAO:
            individuo[i] = 1 - individuo[i]
    return individuo

def algoritmo_genetico() -> tuple[int, float, list[int], bool]:
    """
    Realiza o algoritmo gen√©tico para resolver o problema das oito rainhas.

    Returns:
    -----

    tuple[int, float, list[int], bool]
    """

    populacao = [[random.randint(0, 1) for _ in range(TAMANHO_CROMOSSOMO)] for _ in range(TAMANHO_POPULACAO)]

    inicio = time.time()
    melhor_solucao_encontrada = []

    for geracao in range(MAX_GERACOES):
        notas_aptidao = [calcular_aptidao(ind) for ind in populacao]

        max_aptidao = max(notas_aptidao)
        indice_melhor = notas_aptidao.index(max_aptidao)
        melhor_individuo = populacao[indice_melhor]

        if max_aptidao == 28:
            fim = time.time()
            return geracao + 1, fim - inicio, decodificar_binario(melhor_individuo), True

        nova_populacao = [melhor_individuo[:]]

        while len(nova_populacao) < TAMANHO_POPULACAO:
            pai1 = selecao_roleta(populacao, notas_aptidao)
            pai2 = selecao_roleta(populacao, notas_aptidao)

            filho1, filho2 = cruzamento(pai1, pai2)

            nova_populacao.append(mutacao(filho1))
            if len(nova_populacao) < TAMANHO_POPULACAO:
                nova_populacao.append(mutacao(filho2))

        populacao = nova_populacao

    fim = time.time()

    melhor_final = populacao[notas_aptidao.index(max(notas_aptidao))]
    return MAX_GERACOES, fim - inicio, decodificar_binario(melhor_final), False
```

## **b)Execute o algoritmo 50 vezes e calcule: m√©dia e desvio padr√£o do n√∫mero m√≠nimo de itera√ß√µes necess√°rio para parar o algoritmo; m√©dia e desvio padr√£o do tempo de execu√ß√£o do algoritmo.**

```python
num_execucoes = 50
lista_iteracoes = []
lista_tempos = []
solucoes_unicas = set()

for i in range(num_execucoes):
    iters, tempo, solucao, encontrou_otima = algoritmo_genetico()
    lista_iteracoes.append(iters)
    lista_tempos.append(tempo)

    if encontrou_otima:
        solucoes_unicas.add(tuple(solucao))

media_iter = statistics.mean(lista_iteracoes)
desvio_iter = statistics.stdev(lista_iteracoes)

media_tempo = statistics.mean(lista_tempos)
desvio_tempo = statistics.stdev(lista_tempos)

print("\nResultados:")
print(f"Qauntidade de execu√ß√µes: {num_execucoes}\n")

print(f"M√©dia de Itera√ß√µes: {media_iter:.2f}")
print(f"Desvio Padr√£o das Itera√ß√µes: {desvio_iter:.2f}\n")

print(f"M√©dia de Tempo: {media_tempo:.4f} segundos")
print(f"Desvio Padr√£o do Tempo: {desvio_tempo:.4f} segundos")
```
